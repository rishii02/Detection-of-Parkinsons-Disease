```python
df=pd.read_csv('parkinsons.data')
display(df)
```

```python
df.drop(['name'],axis=1,inplace=True)
display (df)
```

```python
X=df.drop(labels=['status'],axis=1)
Y=df['status']
display(X,Y)
```

```python
X_tr,X_te,Y_tr,Y_te=train_test_split(X,Y,test_size=0.25,random_state=49)
print (X.shape,Y.shape)
print(X_tr.shape,X_te.shape,Y_tr.shape,Y_te.shape)

```

```python
corr = df.corr()
display (corr)
```

```python
log_r = LogisticRegression().fit(X_tr, Y_tr)
train_pre = log_r.predict(X_tr)
print("acc on train : ", accuracy_score(Y_tr, train_pre))
test_pre = log_r.predict(X_te)
print("acc on test : ", accuracy_score(Y_te, test_pre))

```

```python
print(confusion_matrix(Y_tr, train_pre))
print("confusion_matrix test is:\n ",confusion_matrix(Y_te, test_pre))
print('\nClassification Report Train')
print(classification_report (Y_tr, train_pre))
print('\nClassification Report Test')
print(classification_report (Y_te, test_pre))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier().fit(X,Y)
train_pre1=dt.predict(X_tr)
print("train acc:", accuracy_score(Y_tr,train_pre1))
```

```python
test_pre1=dt.predict(X_te)
print("test acc:", accuracy_score(Y_te,test_pre1))
```

```python
print(metrics.cohen_kappa_score(Y_te,test_pre1))
```

```python
print("confusion mat train\n", confusion_matrix(Y_tr,train_pre1))
```

```python
print("confusion mat test\n", confusion_matrix(Y_te,test_pre1))
```

```python
RF=RandomForestClassifier().fit(X_tr,Y_tr)
train_pre2 = RF.predict(X_tr)
print("TRAIN ACC:", accuracy_score(Y_tr,train_pre2))
```

```python
test_pre2=RF.predict(X_te)
print("test acc:",accuracy_score(Y_te,test_pre2))
```

```python
print("confusion mat train\n", confusion_matrix(Y_tr,train_pre2))
```

```python
print("confusion mat test\n", confusion_matrix(Y_te,test_pre2))
```

```python
print(metrics.cohen_kappa_score(Y_te,test_pre2))
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_tr,Y_tr)
train_pre3=nb.predict(X_tr)
test_pre3=nb.predict(X_te)
```
